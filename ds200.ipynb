```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rc('xtick', labelsize='xx-large')
plt.rc('ytick', labelsize='xx-large')
plt.rc('axes', labelsize='xx-large', titlesize='xx-large')
plt.rcParams['legend.title_fontsize'] = 'xx-large'
plt.rcParams['legend.fontsize'] = 'x-large'
```

```python
data = pd.read_csv('data.csv')
```

```python
bihar_data = data[data['SUBDIVISION'] == 'Bihar']
```

```python
# Scatter plot of rains in bihar between 2001 and 2010
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_title('Monthly Rainfall in (2001 - 2010) for Bihar')
ax.set_xlabel('Month')
ax.set_ylabel('Rainfall (mm)')
bihar_data = bihar_data[bihar_data['YEAR'] >= 2001][:10]
scatter_data = bihar_data.iloc[:, 2:14]
x_axis = [1 + i for i in range(12)]
year = 2001
for index, row in scatter_data.iterrows():
    ax.scatter(x_axis, row.to_numpy(), label=year)
    year += 1
ax.grid(which='major', linestyle='-', linewidth=0.5)
ax.grid(which='minor', linestyle='--', linewidth=0.2)
ax.minorticks_on()
ax.set_xticks(x_axis)
ax.set_axisbelow(True)
ax.set_ylim(0)
ax.set_xlim(0)
ax.legend(ncols=4)
plt.savefig('scatter.jpg', dpi=600, bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
# Box plot showing variation of rains in every month for last 10 years
keys = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
box_data = {}
for key in keys:
    box_data[key] = bihar_data[key].to_numpy()
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_title('Variation in Monthly Rainfall (2001 - 2010) for Bihar')
ax.set_xlabel('Month')
ax.set_ylabel('Rainfall (mm)')
ax.boxplot(box_data.values())
ax.grid(which='major', linestyle='-', linewidth=0.5)
ax.grid(which='minor', linestyle='--', linewidth=0.2)
ax.minorticks_on()
ax.set_axisbelow(True)
ax.set_ylim(0)
ax.set_xlim(0)
plt.savefig('box.jpg', dpi=600, bbox_inches='tight', pad_inches=0)
plt.show()   
```

```python
# Bar plot showing variation of rains in every month for 2010
keys = [1 + i for i in range(12)]
bar_data = bihar_data[bihar_data['YEAR'] == 2010].to_numpy()
bar_data = bar_data[0][2:14]
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_title('Variation in Monthly Rainfall 2010 for Bihar')
ax.set_xlabel('Month')
ax.set_ylabel('Rainfall (mm)')
ax.bar(keys, bar_data)
ax.grid(which='major', linestyle='-', linewidth=0.5)
ax.grid(which='minor', linestyle='--', linewidth=0.2)
ax.minorticks_on()
ax.set_xticks(keys)
ax.set_axisbelow(True)
ax.set_ylim(0)
ax.set_xlim(0)
plt.savefig('bar.jpg', dpi=600, bbox_inches='tight', pad_inches=0)
plt.show()   
```
